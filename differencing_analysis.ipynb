```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import kpss
```

```python
def kpss_test(series, **kw):    
    statistic, p_value, n_lags, critical_values = kpss(series, **kw)
    # Format Output
    print(f'KPSS Statistic: {statistic}')
    print(f'p-value: {p_value}')
    print(f'num lags: {n_lags}')
    print('Critial Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
    print(f'Result: The series is {"not " if p_value < 0.05 else ""}stationary')
```

```python
data = pd.read_csv('./aapl_indicators.csv')
data = data.drop(['Unnamed: 0', 'date'], axis=1)
data.head()
```

```python
open_data = data['open']
close_data = data['close']
low_data = data['low']
high_data = data['high']
```

```python
print("Open")
kpss_test(open_data)
print("Close")
kpss_test(close_data)
print("Low")
kpss_test(low_data)
print("High")
kpss_test(high_data)
```

```python
def diff(series):
    l_i = list(series)
    l_o = []
    for i in range(len(l_i)-1):
        l_o.append(l_i[i+1] - l_i[i])
    return(np.array(l_o))
```

```python
diffopen = diff(open_data)
diffclose = diff(close_data)
difflow = diff(low_data)
diffhigh = diff(high_data)
```

```python
print("Open")
kpss_test(diffopen)
print("Close")
kpss_test(diffclose)
print("Low")
kpss_test(difflow)
print("High")
kpss_test(diffhigh)
```

```python
def add_zero(series):
    return(np.array([0] + list(series)))
```

```python
f_open = add_zero(diffopen)
f_close = add_zero(diffclose)
f_high = add_zero(diffhigh)
f_low = add_zero(difflow)
```

```python
fl = [f_open, f_close, f_high, f_low]
l2 = ['open', 'close', 'low', 'high']
for i in range(len(fl)):
    data[l2[i]] = fl[i]
```

```python
diffdata = data.iloc[1:]
```

```python
diffdata.to_csv('./diffaapl.csv')
```
