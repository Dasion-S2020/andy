```python
import anfis
import pandas as pd
import numpy as np
import math
import pywt
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/andyliu/Downloads/TATASTEEL.NS (1).csv')
```

```python
data = df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
y = data['Close']
x = data.drop('Close', axis=1)
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
y = scaler.fit_transform(np.array(y).reshape(-1, 1))
```

```python
x_t = pywt.wavedec(x, wavelet='db2', level=3, mode='symmetric')
```

```python
def getFeatures(x_transformed):
    d = {}
    features = ['Open', 'High', 'Low', 'Volume']
    for i in range(len(x_transformed)):
        feature = features[i]
        a1 = []
        d1 = []
        d2 = []
        for j in range(len(x_transformed[i])):
            a1.append(x_transformed[i][j][0])
            #a1.append(min(x_transformed[i][j][0], 1))
            d1.append(x_transformed[i][j][1])
            #d1.append(min(x_transformed[i][j][1], 1))
            d2.append(x_transformed[i][j][2])
            #d2.append(min(x_transformed[i][j][2], 1))
        l = [a1, d1, d2]
        for k in range(3):
            s = feature + str(k)
            d[s] = l[k]
    return(d)
```

```python
X = getFeatures(x_t)
X
```
