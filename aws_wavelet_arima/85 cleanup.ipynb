```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
pred = np.load('pred.npy')
```

```python
data = pd.read_csv('aapl_indicators_yahoo.csv')['Close']
```

```python
test_n = len(data)*4//5
```

```python
actual = np.array(data)[test_n:]
```

```python
plt.plot(pred, color='r')
plt.plot(actual, color='g')
```

```python
pred1 = pred[:990]
actual1 = actual[:990]
```

```python
plt.plot(pred1, color='r')
plt.plot(actual1, color='g')
plt.savefig('84actual.png')
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return(mape)
```

```python
mse = mean_squared_error(actual1, pred1)
rmse = math.sqrt(mse)
mae = mean_absolute_error(actual1, pred1)
mape = mean_absolute_percentage_error(actual1, pred1)
```

```python
print(mse, rmse, mae, mape)
```
